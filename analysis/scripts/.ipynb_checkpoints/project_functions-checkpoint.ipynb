```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns
```

```python
def load_and_process(url_or_path_to_csv_file):

    # Method Chain 1 (Load data, deal with missing data, sort data, and reset index)

    df1 = (
          pd.read_csv(url_or_path_to_csv_file)
          .dropna()
          .sort_values('alcohol', ascending=False)
          .reset_index(drop=True)
      )

    # Make sure to return the latest dataframe

    return df1
```

```python
df = load_and_process("../../data/raw/winequality-white.csv")
df
```

```python
def rename_and_drop(df):
    
    # Method Chain 2 (Rename columns, drop outliers, and reset the index)

     df2 = (
          df
          .rename(columns={"free sulfur dioxide": "free SO2"})
          .rename(columns={"total sulfur dioxide": "total SO2"})
          #.assign(High_quality = lambda x: x['quality']>3)
          .reset_index(drop=True)
      )

     return df2
```
